```python
# Covert list to numpy array
import numpy as np
a = np.array(a)
b= np.array(b)
c = np.array(c)
inp = np.array(inp)
```

```python
# Calculate the distance between inp and a,b,c using np.linalg.norm
# np.linalg.norm(inp-a) is the distance between inp and a
# Euclidean distance is the square root of the sum of the squared differences between the coordinates of the two points.
dist_inp_a = np.linalg.norm(inp-a)
dist_inp_b = np.linalg.norm(inp-b)
dist_inp_c = np.linalg.norm(inp-c)
```

```python
dist_inp_a , dist_inp_b, dist_inp_c
# Meaning the distance between inp and a is 0.727926605892479, the distance between inp and b is 0.8301453531143942, and the distance between inp and c is 1.2219153747459328
```

```python
closest = min(dist_inp_a, dist_inp_b, dist_inp_c)
print(closest)
```
