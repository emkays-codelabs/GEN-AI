```python
# create a function that take student id as input and returns student name from the dictionary in python 

students = {1: "mk", 2: "vk", 3: "ck"}

def get_student_name(student_id: int):
    return students.get(student_id, "Student not found") # .get() method is used to retrieve the value for the given key (student_id). If the key is not found, it returns "Student not found".
    
    

get_student_name(2)

```

```python
students[2] # accessing value using key directly by indexing
```

```python
for key in students:
    print(f"Key: {key}, Value: {students[key]}")    
```

```python
for key, value in students.items():
    print(f"Key: {key}, Value: {value}")
```
